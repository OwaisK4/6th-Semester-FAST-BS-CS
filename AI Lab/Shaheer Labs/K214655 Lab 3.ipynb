4. Perform the following tasks with NLTK.
    - Write a Python NLTK program to split the text sentence/paragraph into a list of words.

String = Joe waited for the train. The train was late. Mary and Samantha took the bus. I looked for Mary and Samantha at the bus station.

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

text = """Joe waited for the train. The train was late. Mary and Samantha took the bus. I looked for Mary and Samantha at the bus station."""

# Tokenize text into words
word_tokens = word_tokenize(text)
print("Word tokens:")
print(word_tokens)

# Tokenize text into sentences
sentences = sent_tokenize(text)
print("\nSentences:")
for sentence in sentences:
    print(sentence)
```

```python
data = [
    ['D', 'D', 'D', 'D', 'D', 'D', 'D', 'B'],
    ['B', 'B', 'C', 'D', 'B', 'B', 'B', 'B'],
    ['D', 'D', 'C', 'D', 'B', 'D', 'D', 'D'],
    ['D', 'D', 'D', 'D', 'C', 'C', 'C', 'D'],
    ['D', 'D', 'D', 'D', 'D', 'D', 'D', 'D'],
    ['D', 'D', 'D', 'D', 'D', 'D', 'D', 'D'],
    ['D', 'D', 'D', 'B', 'B', 'D', 'D', 'D']
]

for row in data:
    for x in row:
        if(x == 'B'):
            print(''+ ' ', end='')
        else:
            print(x + ' ' , end='') 
    print()

start = (0,0)
i,j = start


while True:
    if data[i][j] == 'D':
        data[i][j] = 'C'
    elif i < len(data) - 1 and data[i + 1][j] == 'D': #Down
        i+=1
        data[i][j] = 'C'
    elif j > 0 and data[i][j - 1] == 'D': #Left
        j-=1
        data[i][j] = 'C'
    elif i > 0 and data[i - 1][j] == 'D': #Up
        i-=1
        data[i][j] = 'C'
    elif j < len(data[0]) - 1 and data[i][j + 1] == 'D': #Right
        j+=1
        data[i][j] = 'C'
    else:
        break

print()
for row in data:
    for x in row:
        if(x == 'B'):
            print(''+ ' ', end='')
        else:
            print(x + ' ' , end='') 
    print()
    
print(f"\nEnd i: {i}, j: {j}")
```

empty spaces of the grid. The goal of the computer agent is to win the game. You can assign (X or O)

game. If there is no winning move on any strike then the computer marks anywhere on the board.

is an algorithm that calculates a path between two nodes in a weighted graph such as the sum of the values
on the edges that form a path is minimized. Starting from the source node, the algorithm looks up the
weights on the (out-)going (in weighted graphs) edges. It chooses the edge which, summed to the
previous total sum, gives the lowest result. The algorithm runs through every node up until the destination
point. Results are a path and the total sum of the shortest path. Your task is to write a program of utility-
based agent that find the best route (minimum Distance, shortest path) from source to destination node.

```python
class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0 for _ in range(vertices)] for _ in range(vertices)]

    def printSolution(self, dist):
        print(f"Vertex \t\t Distance from {places[src]}")
        for node in range(self.V):
            print(f"{places[node]} ({node}) \t {dist[node]}")

    def minDistance(self, dist, visited):

        min = float("inf")
        for v in range(self.V):
            if dist[v] < min and visited[v] == False:
                min = dist[v]
                min_index = v

        return min_index

    def dijkstra(self, src):

        dist = [float("inf")] * self.V
        dist[src] = 0
        visited = [False] * self.V

        for _ in range(self.V):

            u = self.minDistance(dist, visited)
            visited[u] = True
            for v in range(self.V):
                if (
                    self.graph[u][v] > 0
                    and visited[v] == False
                    and dist[v] > dist[u] + self.graph[u][v]
                ):
                    dist[v] = dist[u] + self.graph[u][v]

        self.printSolution(dist)


places = ["Fast", "Karsaz", "Gulshan", "Korangi", "Saddar"]
g = Graph(5)
g.graph = [
    [0, 15, 30, 20, 0],  # Fast
    [15, 0, 8, 10, 10],  # Karsaz
    [30, 8, 0, 0, 20],  # Gulshan
    [20, 10, 0, 0, 10],  # Korangi
    [0, 10, 20, 10, 0],  # Saddar
]
src = 0
g.dijkstra(src)
```
